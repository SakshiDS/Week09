### a. In a jupyter notebook markdown cell, describe the difference between linear and rank correlation.

Linear correlation is the correlation coeffcient between the values of two variables and analysies how close are the relationship to being positively or negatively linear.

Rank correlation is the correlation coeffcient between the ranks of the values of the 2 variables. It analyses the how closely linear are the values ordered in the dataset.

### b. Create a scatter plot showing the relationship between calculated_host_listings_count and availability_365 in the Belize Airbnb data

```python
fig, ax = plt.subplots()

ax.scatter(belize_df['calculated_host_listings_count'], belize_df['availability_365'], color= 'blue')

# Set the x-axis label
ax.set_xlabel('Calculated Host Listings Count')

# Set the y-axis label 
ax.set_ylabel('Availability_365')

# Set the title
ax.set_title("Relationship between Calculated Host Listings Count & Availability 365")

plt.grid()


plt.show()
```

### c. Calculate the correlation coefficient for the data in the scatter plot, and describe what is happening

```python
np.corrcoef(belize_df['calculated_host_listings_count'],belize_df['availability_365'])
```

The correlation coefficient of the data is 0.17 which is a weak to negligible relationship.

It means that 'calculated_host_listings_count' & 'Availability_365' hardly impact each other.

## 2. Create a box and whisker plot for the Belize Airbnb data price. In a markdown cell, describe what is going on in as simple of terms as possible. Use the documentation
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
ax = belize_df[["price"]].boxplot(vert=False, figsize=(15,5))

plt.title("Box plot for price value disribution")
plt.show()
```

The box plot is skewed because of way too many outliers far and above the upper quartile.
While there are no outliers below the lower quartile.

In the plot below, the outliers are ignored and that provides a better understanding of price distribution.

```python
ax = belize_df[["price"]].boxplot(vert=False,sym="", figsize=(15,5))

plt.title("Box plot for price value disribution")
plt.show()
```
